```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tested_data.csv", sep=';')
data
```

```python
random_arrs = data[data["array_type"] == "random"]
random_arrs.head(10)
```

```python
quicks = random_arrs[random_arrs["sorting_type"] == "quick"]
intros = random_arrs[random_arrs["sorting_type"] == "intro"]
```

```python
plt.plot(quicks['n'], quicks['time'], label="quick sort")
plt.plot(intros["n"], intros['time'], label="intro sort")

plt.legend()
plt.grid()

ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.title("Time for sorting random arrays",{'family':'serif','color':'blue','size':20})
plt.xlabel("Lenght of array")
plt.ylabel("nanoseconds")
```

```python
reversed_arrs = data[data["array_type"] == "reversed"]
reversed_arrs.head(10)
```

```python
quicks = reversed_arrs[reversed_arrs["sorting_type"] == "quick"]
intros = reversed_arrs[reversed_arrs["sorting_type"] == "intro"]
```
