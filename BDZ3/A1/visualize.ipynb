```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_corr = pd.read_csv('n_correlation.csv', sep = ';')
n_corr
```

```python
plt.plot(n_corr['n'], n_corr['area'])

ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.title("Predicted area correlation for n",{'family':'serif','color':'blue','size':20})
plt.xlabel("N")
plt.ylabel("predicted area")
```

```python
plt.plot(n_corr['n'], n_corr['area_diff'])

ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

plt.title("Predicted area difference by n",{'family':'serif','color':'blue','size':20})
plt.xlabel("N")
plt.ylabel("difference with correct area")
```
