```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tested_data.csv", sep=';')
data
```

```python
random_arrs = data[data["type"]=="random"]
random_arrs
```

```python
threshold0 = random_arrs[random_arrs["threshold"] == 0]
threshold5 = random_arrs[random_arrs["threshold"] == 5]
threshold10 = random_arrs[random_arrs["threshold"] == 10]
threshold20 = random_arrs[random_arrs["threshold"] == 20]
threshold30 = random_arrs[random_arrs["threshold"] == 30]
threshold50  =random_arrs[random_arrs["threshold"] == 50]
```

```python
plt.plot(threshold0['n'], threshold0['msec'], label="merge sort")
plt.plot(threshold5['n'], threshold5['msec'], label="merge+insertion 5")
plt.plot(threshold10['n'], threshold10['msec'], label="merge+insertion 10")
plt.plot(threshold20['n'], threshold20['msec'], label="merge+insertion 20")
plt.plot(threshold30['n'], threshold30['msec'], label="merge+insertion 30")
plt.plot(threshold50['n'], threshold50['msec'], label="merge+insertion 50")
plt.legend()
plt.grid()

ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.title("Time for sorting random arrays",{'family':'serif','color':'blue','size':20})
plt.xlabel("Lenght of array")
plt.ylabel("nanoseconds")
```
